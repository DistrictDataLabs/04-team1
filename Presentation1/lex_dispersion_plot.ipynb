```python
import nltk
from nltk.corpus import PlaintextCorpusReader
from nltk.text import Text
import matplotlib
import numpy
import os
import fnmatch
import re
```

```python
# download folder of txt files from DDL/Url_Text
```

```python
txts_dir = "/Users/lisacombs/Desktop/posts_04062016"

docs = []

for dirpath, dirs, files in os.walk(txts_dir):
    for filename in fnmatch.filter(files, '*.txt'):
        with open(os.path.join(dirpath, filename)) as f:
            docs.append(f.read())
```

```python
stop_words = set([ 
    'he','how','in','just', 'like', 'my', 'had', 'his', 'said' , 'she', 'so', 'than', 'their', 'we', 'were', 'what', 
        'which', 'who', 'would', 'her', 'into', 'there', 'youre', 'your', 'also', 'been', 'could', 'do', 'our', 
        'he', 'them', 'me', 'to', 'is', 'by'
])
```

```python
def clean_text(txt):    
    clean_words = []
    for w in txt.split():
        if w not in stop_words and re.search(r"^[A-Za-z].*[A-Za-z]$", w):
            w = re.sub(r"[^A-Za-z-]", "", w)
            if len(w) > 1:
                clean_words.append(w.lower())
    return clean_words
```

```python
texts = [clean_text(d) for d in docs]
```

```python
test = [w for sublist in texts for w in sublist]
```

```python
%matplotlib inline
matplotlib.rcParams["figure.figsize"] = 20, 8
```

```python
from nltk.draw.dispersion import dispersion_plot
dispersion_plot(test, ["clinton", "sanders"])

dispersion_plot(test, ["cruz", "kasich", "trump", "rubio", "bush", "carson"])
```
